# *04. 문자열/슬라이싱/문자열처리함수,포맷/탈출문자*

```python
# 04-1 문자열
sentence = '나는 소년입니다'
print(sentence)
sentence2 = "파이썬은 쉬워요"
print(sentence2)
sentence3 = """ 
나는 소년이고,
파이썬은 쉬워요
"""
print(sentence3)      # 변수를 선언할 때 큰따옴표나 작은따옴표 3개로 시작하고 3개로 끝내면 그 사이에 있는 문자열의 줄바꿈까지 입력된다.


# 04-2. 슬라이싱
jumin = '990120-1234567'

print('성별 : ' + jumin[7])  # 컴퓨터는 첫번째 자리를 1번째가 아닌 0번째라고 함
print('연 : ' + jumin[0:2])  # 콜론의 끝 값은 포함되지 않는다(항상 +1로 입력)
print('월 : ' + jumin[2:4])  # 2 부터 4 직전까지 (2,3)
print('일 : ' + jumin[4:6])

print('생년월일 : ' + jumin[:6]) # 처음부터 6 직전까지
print('뒤 7자리 : ' + jumin[7:]) # 7 부터 끝까지
print('뒤 7자리 (뒤에부터) : ' + jumin[-7:]) # '맨 뒤에서 7번째'부터 끝까지


# 04-3. 문자열 처리 함수
python = 'Python is Amazing'
print(python.lower()) # 소문자 처리
print(python.upper()) # 대문자 처리
print(python[0].isupper()) # 대문자인지 확인
print(len(python))
print(python.replace('Python', 'Java')) # Python 이라는 글자를 찾아 Java로 변경

index = python.index('n') # n 이라는 글자가 몇 번째에 위치하는지
print(index)
index = python.index('n', index + 1) # 위의 index 다음 번째부터 검색
print(index)

print(python.find('n')) # find 함수와 index 함수에는 조금 차이가 있음
print(python.find('Java')) # 내가 원하는 값이 없을 때 find 함수는 -1 을 반환해줌
# print(python.index('Java')) # 내가 원하는 값이 없을 때 index 함수는 오류남
print('hi') # 오류가 나면 코드가 더 이상 진행되지 않음

print(python.count('n')) # 몇 번 등장하는지


# 04-4. 문자열 포맷
# print('a' + 'b')
# print('a', 'b')

# 방법 1
print('나는 %d살입니다.' % 20) # %d는 정수를 의미하며, 이 자리에 뒤의 숫자를 넣겠다는 것임
print('나는 %s을 좋아해요.' % '파이썬') # %s는 문자열을 의미하며, 이 자리의 뒤의 문자열을 넣겠다는 것임
print('Apple은 %c로 시작해요' % 'A') # %c는 한 글자만 받겠다는 것임
# %s는 모든 것을 다 받아들이기 때문에 %s 쓰는 것이 가장 편함
print('나는 %s살입니다.' % 20)
print('나는 %s색과 %s색을 좋아해요.' % ('파란', '빨간'))

# 방법 2
print('나는 {}살입니다.'.format(20)) # 뒤의 값을 중괄호에 넣겠다는 뜻
print('나는 {}색과 {}색을 좋아해요.'.format('파란', '빨간'))
print('나는 {0}색과 {1}색을 좋아해요.'.format('파란', '빨간')) # 중괄호에 순서를 부여하는 것
print('나는 {1}색과 {0}색을 좋아해요.'.format('파란', '빨간'))

# 방법 3
print('나는 {age}살이며, {color}색을 좋아해요.'.format(age = 20, color = '빨간')) # 중괄호에 변수 넣는 것
print('나는 {age}살이며, {color}색을 좋아해요.'.format(color = '빨간', age = 20))

# 방법 4 (python v3.6 이상)
age = 20
color = '빨간'
print(f'나는 {age}살이며, {color}색을 좋아해요.')


# 04-5. 탈출문자
# \n : 줄바꿈
'''백문이 불여일견
백견이 불여일타'''
print('백문이 불여일견 백견이 \n불여일타')

# \" 또는 \' : 문장 내에서 따옴표
# 저는 '나도코딩'입니다
# print('저는 '나도코딩'입니다.') 오류가 남
print('저는 "나도코딩"입니다.') # 큰따옴표, 작은따옴표 따로 쓰면 출력 가능
print('저는 \'나도코딩\'입니다.') 

# \\ : 문장 내에서 \
print('D:\\PythonWorkspace')

# \r : 커서를 맨 앞으로 이동
print('Red Apple\rPine') # 커서를 맨 앞으로 이동시킨 후 앞의 4자리를 Pine로 채우는 것

# \b : 백스페이스 (한 글자 삭제)
print('Redd\bApple') # d 하나를 삭제

# \t : 탭
print('Red\tApple')


# 04-6. Quiz3) 사이트별로 비밀번호를 만들어 주는 프로그램을 작성하시오

# 예) http://naver.com
# 규칙1 : http:// 부분은 제외 => naver.com
# 규칙2 : 처음 만나는 점(.) 이후 부분은 제외 => naver
# 규칙3 : 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e' 갯수 + '!' 로 구성
#                    (nav)            (5)             (1)        (!)
# 예) 생성된 비밀번호 : nav51!

# 내 풀이
site = 'http://naver.com'
password = site[7:10] + str(len(site[7:-4])) + str(site.count('e')) + '!'
print(password)

# 나도코딩 풀이
# url = 'http://google.com'
# url = 'http://daum.net'
# url = 'http://naver.com'
url = 'http://youtube.com'
my_str = url.replace('http://','') # 규칙1 적용 => http:// 부분을 빈칸으로 바꾸기
# print(my_str)
my_str = my_str[:my_str.index('.')] # 규칙2 적용 => 처음부터 점 나오기 전까지만 사용
# print(my_str)
password = my_str[:3] + str(len(my_str)) + str(my_str.count('e')) + '!'
print('{0} 의 비밀번호는 {1} 입니다.'.format(url, password))
```
